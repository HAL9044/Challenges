```python
import pandas as pd
```

```python
df = pd.read_csv('./data/input/IMDB-Movie-Data.csv')
```

```python
def vote_counter(votes):
    if votes < 999:
        return 1
    elif votes >= 1000 and votes < 9999:
        return 2
    elif votes >= 10000 and votes < 99999:
        return 3
    elif votes >= 100000 and votes < 999999:
        return 4
    elif votes > 1000000:
        return 5   
```

```python
df.loc[:,"Votes"]
```

```python
df["Bins"] = df["Votes"].apply(vote_counter)
df
```

```python
def rvn_min (rev):
    return rev[0] / rev[1]
```

```python
df['Rvn_Min'] = df[["Revenue (Millions)", "Runtime (Minutes)"]].apply(rvn_min, axis=1)
```

```python
def rvn_min (rev, min):
    return rev / min
```

```python
df['Rvn_Min'] = df.apply(lambda x: rvn_min(x["Revenue (Millions)"], x["Runtime (Minutes)"]), axis=1 )
```

```python
def Trueif (str, rank):
    if "Thriller" in str:
        return rank + 1
    elif "Comedy" in str:
        return rank -1
    else: return rank
```

```python
df["NewRanking"] = df.apply(lambda x: Trueif(x["Genre"], x["Rating"]), axis = 1)
```

```python
#Challenge 4
def ASCII (str, votes):
    sum = 0
    
    for char in str:
        sum = sum + ord(char)
    
    int_div = int(sum/votes)
    
    if int_div == 1:
        return True
    elif sum < votes:
        return False
    else: 
        for i in range (2, int_div - 1):
            if int_div % i == 0:
                return False
            else: 
                return True
```

```python
df["Prime?"] = df.apply(lambda x: ASCII(x["Title"], x["Votes"]), axis = 1)    
```

```python
def Rvn_Rank (rvn, rank, meta):
    avg = (rank + (meta/10)) / 2
    sc = rvn / avg
    return sc
```

```python
df["Avg_Sc"] = df.apply(lambda x: Rvn_Rank(x["Revenue (Millions)"], x["Rating"], x["Metascore"]), axis = 1)
df
```
